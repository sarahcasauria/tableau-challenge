```python
# Import dependencies
import pandas as pd
import numpy as np
import datetime as dt
```

```python
# Define two rounding functions that will be used to calculate trip time later on
def round_half_up_2dec(n, decimals=2):
    multiplier = 10 ** decimals
    return np.floor(n*multiplier + 0.5) / multiplier
```

```python
# Load all 2020 files
jan_2020 = pd.read_csv("raw_data/JC-202001-citibike-tripdata.csv")
feb_2020 = pd.read_csv("raw_data/JC-202002-citibike-tripdata.csv")
march_2020 = pd.read_csv("raw_data/JC-202003-citibike-tripdata.csv")
april_2020 = pd.read_csv("raw_data/JC-202004-citibike-tripdata.csv")
may_2020 = pd.read_csv("raw_data/JC-202005-citibike-tripdata.csv")
june_2020 = pd.read_csv("raw_data/JC-202006-citibike-tripdata.csv")
july_2020 = pd.read_csv("raw_data/JC-202007-citibike-tripdata.csv")
august_2020 = pd.read_csv("raw_data/JC-202008-citibike-tripdata.csv")
september_2020 = pd.read_csv("raw_data/JC-202009-citibike-tripdata.csv")
october_2020 = pd.read_csv("raw_data/JC-202010-citibike-tripdata.csv")
november_2020 = pd.read_csv("raw_data/JC-202011-citibike-tripdata.csv")
december_2020 = pd.read_csv("raw_data/JC-202012-citibike-tripdata.csv")
```

```python
# Combine DataFrame objects with overlapping columns and 
# return only those that are shared by passing inner to the join keyword argument.
all_2020 = pd.concat([jan_2020, 
                  feb_2020, 
                  march_2020, 
                  april_2020, 
                  may_2020, 
                  june_2020, 
                  july_2020, 
                  august_2020, 
                  september_2020,
                  october_2020,
                  november_2020,
                  december_2020], ignore_index=True)
all_2020
```

```python
# Rename columns
all_2020.rename(columns={"tripduration":"Trip Duration (sec)",
                         "starttime":"Start Time",
                         "stoptime":"Stop Time",
                         "start station name":"Start Station Name",
                         "start station latitude":"Start Lat",
                         "start station longitude":"Start Lon",
                         "end station name":"End Station Name",
                         "end station latitude":"End Lat",
                         "end station longitude":"End Lon",
                         "usertype":"User Type"}, inplace=True)

all_2020
```

```python
# Remove any NaN values for a cleaner dataset
all_2020.dropna(inplace=True)
all_2020
```

```python
# Remove unecessary columns (want to be able to make comparisons across multiple years of data and rider demographics not
# present in later years.)

all_2020 = all_2020[["Start Time",
                     "Stop Time",
                     "Start Station Name",
                     "Start Lat",
                     "Start Lon",
                     "End Station Name",
                     "End Lat",
                     "End Lon",
                     "User Type",
                     "Trip Duration (sec)"]]

all_2020
```

```python
# Convert start and stop times to datetime (y-m-d h:s:ms)
all_2020['Start Time'] = pd.to_datetime(all_2020['Start Time'])
all_2020["Stop Time"] = pd.to_datetime(all_2020['Stop Time'])

all_2020.dtypes
```

```python
# Create two new columns, one for trip duration in mins and one in hrs (better to do this now than in Tableau)
all_2020['Trip Duration (min)'] = round_half_up_2dec((all_2020["Stop Time"]-all_2020['Start Time'])/pd.Timedelta(minutes=1))
all_2020['Trip Duration (hrs)'] = round_half_up_2dec((all_2020["Stop Time"]-all_2020['Start Time'])/pd.Timedelta(hours=1))

all_2020
```

```python
# Export to .csv
all_2020.index.rename("Ride ID", inplace=True)

all_2020.to_csv("cleaned_data/rides_2020.csv")
```
