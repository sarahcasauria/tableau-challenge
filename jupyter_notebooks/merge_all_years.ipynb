```python
# Import dependencies
import pandas as pd
```

```python
# Read in all three dataframes
df_2019 = pd.read_csv("cleaned_data/rides_2019.csv")
df_2020 = pd.read_csv("cleaned_data/rides_2020.csv")
df_2021 = pd.read_csv("cleaned_data/rides_2021.csv")
```

```python
# list all column names to ensure they are the same
dfs=[df_2019, df_2020, df_2021]

for df in dfs:
    print(df.columns.tolist())
```

```python
# Merge all dataframes together
all_data = pd.concat([df_2019, df_2020, df_2021])

all_data
```

```python
# Remove any rows where trip duration is less than 1 minute
all_data = all_data.loc[all_data["Trip Duration (min)"]>=1,:]
all_data
```

```python
# Remove any rows where lat or lon is 0
all_data = all_data.loc[(all_data["End Lat"]!=0) & 
                        (all_data["Start Lat"]!=0) & 
                        (all_data["End Lon"]!=0) & 
                        (all_data["Start Lon"]!=0),:]

all_data
```

```python
# Export to CSV
all_data.to_csv("cleaned_data/all_data.csv", index=False)
```
